# Central Limit Theorem 

# Formula 

### z = (x-u)/standard_deviation/squareroot(n)

### t = (t-nu)/squareroot(n)*Standard Deviation

Question 1: Let X be a random variable along with µ = 10 and σ = 4. A sample of size 100. Find the probability that the sample mean ‘µ’ of 100 number of observations is not more than (less than) 9. 

```python
u = 10
st_dev = 4
n = 100
x = 9
#T Statistics will be equal to
#p(x<9) = p(z<(x-u)/(st_dev/math.sqrt(n)))
val_ = (x-u)/(st_dev/math.sqrt(n))
val_
print('Z socre of',val_, 'is equal to', 0.00621)
```

Question 3: A large freight elevator can transport a maximum of 9800 pounds. Suppose a load of cargo containing 49 boxes must be carried through the elevator. Experience has shown that the weight of some boxes of this type of cargo follows a distribution with mean µ = 205 pounds & the standard deviation ‘σ’ = 15 pounds. Based on this information, what is a probability that all 49 boxes can be safely loaded onto the freight elevator and transported?

```python

u = 205
st_dev = 15
n = 49
t = 9800

a = (t)-(n*u)
a
b = (math.sqrt(n) * st_dev)
b

t_val = a/b
print('Z socre of',t_val, 'is equal to', 0.00990)
```

Question 4: It is known that the number of tickets purchased by a student standing in line at the ticket counter to buy the tickets for the Football match of U.C.L.A. against USC follows a distribution that has a mean µ = 2.4 & the standard deviation ‘σ’ = 2.0. 
Suppose 100 eager students are standing in line to purchase the tickets. If only 250 tickets remain unbooked, what is a Probability that all 100 students will be able to buy the tickets they desire?


```python

u = 2.4
st_dev = 2.0
n = 100
t = 250

a = (t)-(n*u)
a
b = (math.sqrt(n) * st_dev)
b

t_val = a/b
print('Z socre of',t_val, 'is equal to', 0.6915)
```
